```python
from mlflow.tracking import MlflowClient


MLFLOW_TRACKING_URI = "sqlite:///mlflow.db"
```

```python
client = MlflowClient(tracking_uri=MLFLOW_TRACKING_URI)
```

```python
from mlflow.entities import ViewType

runs = client.search_runs(
    experiment_ids="1",
    filter_string="tags.mlflow.runName = 'silent-ram-690'",
    run_view_type=ViewType.ACTIVE_ONLY,
    max_results=5,
)
```

```python
import mlflow
```

```python
run_id = runs[0].info.run_id
model_uri = f"runs:/{run_id}/mlflow_model"
mlflow.register_model(model_uri=model_uri, name="nyc-taxi-ridge")
```

```python
client.set_registered_model_alias(name="nyc-taxi-ridge", alias="staging", version=2)
```

```python
client.get_registered_model(name="nyc-taxi-ridge")
```

```python
client.get_latest_versions(name="nyc-taxi-ridge")[0].version
```

```python
model_version = 2
model_name = "nyc-taxi-ridge"
new_stage = "Staging"
client.transition_model_version_stage(
    name=model_name,
    version=model_version,
    stage=new_stage,
    archive_existing_versions=False,
)
```

```python
from datetime import datetime

date = datetime.today().date()
client.update_model_version(
    name=model_name,
    version=model_version,
    description=f"The model version {model_version} was transitioned to {new_stage} on {date}",
)
```
